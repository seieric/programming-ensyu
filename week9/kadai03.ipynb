```python
from random import random
import statistics

class RandomWalk():
    def __init__(self, p, x0):
        if p < 0 or p > 1:
            raise ValueError("pは0以上1以下である必要があります")
        self.p = p
        self.x = x0

    def next_step(self):
        if (random() <= self.p):
            self.x += 1
        else:
            self.x -= 1

    def gen_traj(self, num_steps):
        traj = []
        for _ in range(num_steps):
            self.next_step()
            traj.append(self.x)
        return traj

    def calc_stats(self, num_steps, num_trials):
        pos = []
        for _ in range(num_trials):
            rw = RandomWalk(self.p, self.x)
            for _ in range(num_steps):
                rw.next_step()
            pos.append(rw.x)

        mean = statistics.mean(pos)
        var = statistics.variance(pos)
        return mean, var
```

```python
import matplotlib.pyplot as plt

rw = RandomWalk(0.5, 0)
traj = rw.gen_traj(1000)

plt.plot(traj)
plt.show()
```

```python
mean, var = rw.calc_stats(100, 10000)
print("mean: ", mean)
print("var: ", var)
```
