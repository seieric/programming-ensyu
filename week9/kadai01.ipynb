```python
class Matrix:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2

    def __repr__(self):
        s = f"[{self.x1}, {self.y1}]\n"
        s += f"[{self.x2}, {self.y2}]"
        return s

    def add(self, mat):
        x1 = self.x1 + mat.x1
        y1 = self.y1 + mat.y1
        x2 = self.x2 + mat.x2
        y2 = self.y2 + mat.y2
        return Matrix(x1, y1, x2, y2)

    def mul(self, mat):
        x1 = self.x1 * mat.x1 + self.y1 * mat.x2
        y1 = self.x1 * mat.y1 + self.y1 * mat.y2
        x2 = self.x2 * mat.x1 + self.y2 * mat.x2
        y2 = self.x2 * mat.y1 + self.y2 * mat.y2
        return Matrix(x1, y1, x2, y2)
```

```python
# 和
m1 = Matrix(1,2,5,1)
m2 = Matrix(1,0,0,1)
m3 = m1.add(m2)
print(m3)
print("------")
# 積 (m1*m2)
m4 = m1.mul(m2)
print(m4)
```
