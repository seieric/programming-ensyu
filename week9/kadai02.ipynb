```python
class Matrix:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2

    def __repr__(self):
        s = f"[{self.x1}, {self.y1}]\n"
        s += f"[{self.x2}, {self.y2}]"
        return s

    def add(self, mat):
        x1 = self.x1 + mat.x1
        y1 = self.y1 + mat.y1
        x2 = self.x2 + mat.x2
        y2 = self.y2 + mat.y2
        return Matrix(x1, y1, x2, y2)

    def mul(self, mat):
        x1 = self.x1 * mat.x1 + self.y1 * mat.x2
        y1 = self.x1 * mat.y1 + self.y1 * mat.y2
        x2 = self.x2 * mat.x1 + self.y2 * mat.x2
        y2 = self.x2 * mat.y1 + self.y2 * mat.y2
        return Matrix(x1, y1, x2, y2)

    def div(self, num):
        x1 = self.x1 / num
        y1 = self.y1 / num
        x2 = self.x2 / num
        y2 = self.y2 / num
        return Matrix(x1, y1, x2, y2)
```

```python
x = Matrix(1, 2, 3, 4)
sum = Matrix(0, 0, 0, 0)
# k = 0
m = Matrix(1, 0, 0, 1)
sum = sum.add(m)
# k = 1, 2, ...
for k in range(1, 10000):
    m = m.mul(x.div(k))
    sum = sum.add(m)

print(sum)
```
