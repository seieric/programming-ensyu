```python
import numpy as np

n = 1000

arr = np.random.randn(n, n)
arr /= np.sqrt(n) # 分散を1/nにする
m = np.asmatrix(arr)
eig_val, _  = np.linalg.eig(m)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

xx = [x.real for x in eig_val]
yy = [x.imag for x in eig_val]
plt.scatter(xx, yy)
```
