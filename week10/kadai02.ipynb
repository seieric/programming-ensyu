```python
class MyRational:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __str__(self):
        if (self.a % self.b == 0):
            return f"{self.a // self.b}"
        else:
            return f"{self.a} / {self.b}"

    def __add__(self, other):
        a = self.a * other.b + self.b * other.a
        b = self.b * other.b
        return MyRational(a, b)

    def __mul__(self, other):
        a = self.a * other.a
        b = self.b * other.b
        return MyRational(a, b)

    def __eq__(self, other):
        return self.a / self.b == other.a / other.b

    def __hash__(self):
        return self.a * self.b + self.a
```

```python
import math
class MyInteger(MyRational):
    def __init__(self, a):
        super().__init__(a, 1)

    def factorial(self):
        if (self.a == 0 or self.a == 1):
            return 1
        else:
            return self.a * MyInteger(self.a - 1).factorial()

    def is_prime(self):
        if self.a < 2:
            return False
        for i in range(2, int(math.sqrt(self.a))):
            if (self.a % i == 0):
                return False
        return True
```

```python
i1 = MyInteger(7)
i2 = MyInteger(20)
print(i1 + i2)
print(i1 * i2)
print(i1 == i2)
print(i1.factorial())
print(i1.is_prime())
print(i2.is_prime())
```
