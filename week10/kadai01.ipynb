```python
class MyRational:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __str__(self):
        if (self.a % self.b == 0):
            return f"{self.a // self.b}"
        else:
            return f"{self.a} / {self.b}"

    def __add__(self, other):
        a = self.a * other.b + self.b * other.a
        b = self.b * other.b
        return MyRational(a, b)

    def __mul__(self, other):
        a = self.a * other.a
        b = self.b * other.b
        return MyRational(a, b)

    def __eq__(self, other):
        return self.a / self.b == other.a / other.b

    def __hash__(self):
        return self.a * self.b + self.a
```

```python
r1 = MyRational(2, 3)
print(r1)
r2 = MyRational(1, 3)
print(r1 + r2)
```

```python
r1 = MyRational(1, 3)
r2 = MyRational(1, 3)
r1 == r2
```
