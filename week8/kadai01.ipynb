```python
def logistic_map(a = 2, x0 = 0.5):
    if (a < 0 or 4 < a):
        raise ValueError("aは0以上4以下である必要があります")
    if (x0 < 0 or 1 < x0):
        raise ValueError("x0は0以上1以下である必要があります")
    prev = x0;
    while True:
        yield prev
        prev = a * prev * (1 - prev)
```

```python
import matplotlib.pyplot as plt

g = logistic_map(a = 3.9, x0 = 0.1)
traj = [next(g) for i in range(100)]
plt.plot(traj)
```
