```python
import matplotlib.pyplot as plt
```

```python
def logistic_map(a = 2, x0 = 0.5):
    if (a < 0 or 4 < a):
        raise ValueError("aは0以上4以下である必要があります")
    if (x0 < 0 or 1 < x0):
        raise ValueError("x0は0以上1以下である必要があります")
    prev = x0;
    while True:
        yield prev
        prev = a * prev * (1 - prev)
```

```python
def draw_fig(x0 = 0.1, a0 = 0, a1 = 4):
    xs = []
    ys = []
    for i in range(1000):
        a = i * (a1 - a0) * 0.001 + a0
        g = logistic_map(a = a, x0 = x0)
        # 最初の50個の値を捨てる
        for j in range(50):
            next(g)
        xs.extend([a for n in range(50)])
        ys.extend([next(g) for n in range(50)])
    plt.scatter(xs, ys, s=0.1)
```

```python
draw_fig()
```

```python
draw_fig(a0 = 3.5)
```
