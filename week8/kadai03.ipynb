```python
def rand_val():
    x = 1
    a = 1664525
    c = 1013904223
    m = 4294967296
    while True:
        xnext = (a * x + c) % m
        yield xnext / m
        x = xnext
```

```python
def random_walk(x0 = 0, p = 0.5):
    rand_g = rand_val()
    if p < 0 or p > 1:
        raise ValueError("pは0以上1以下である必要があります")
    while True:
        yield x0
        if next(rand_g) <= p:
            x0 += 1
        else:
            x0 -= 1
```

```python
import matplotlib.pyplot as plt
walk_g = random_walk()
plt.plot([next(walk_g) for _ in range(1000)])
```

```python
walk_g = random_walk(p = 10)
# pの値がおかしいのでエラーになる
next(walk_g)
```
