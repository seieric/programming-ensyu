```python
%%file mylib/nintegrate.py
def nintegrate(f, start, end, num_split):
    dx = (end - start) / num_split
    x = start
    res = 0
    while (x + dx <= end):
        res += (f(x) + f(x + dx)) * dx * 0.5
        x += dx
    return res
```

```python
from mylib.nintegrate import nintegrate
```

```python
nintegrate(lambda x: 1, 0, 10, 10000)
```

```python
nintegrate(lambda x: x, 0, 10, 10000)
```

```python
import math
nintegrate(math.sin, 0, 2*math.pi, 10000)
```

```python
nintegrate(lambda x: x+x**2, 0, 10, 10000)
```
