```python
def nestlist(f, x0, times):
    if times <= 0:
        raise ValueError("timesは1以上の整数値を指定してください")
    results = []
    x = x0
    for i in range(times):
        x = f(x)
        results.append(x)
    return results
```

```python
nestlist(lambda x: x**2, 2, 3)
```

```python
# ランダムウォーク
import random
import matplotlib.pyplot as plt
l = nestlist(
lambda x: x + (-1)**random.randint(0, 1),
0, 1000
)
plt.plot(l)
plt.show()
```

```python
# Tinkerbell写像
a = 0.9
b = -0.6013
c = 2.0
d = 0.5
p0 = (-0.72, -0.64)
l1 = nestlist(
    lambda p: (p[0]**2 - p[1]**2 + a*p[0] + b*p[1], 2*p[0]*p[1] + c*p[0] + d*p[1]),
    p0, 5000
)
x, y = zip(*l1)
plt.scatter(x, y)
plt.show()
```
