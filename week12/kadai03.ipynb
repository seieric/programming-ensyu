```python
def ndsolve(fs, x0, t_start, t_end, dt = 0.01):
    traj = [[t_start, *x0]]

    traj_i = 0
    t = t_start + dt
    while t + dt <= t_end:
        x = [t]
        for i, f in enumerate(fs):
            x.append(traj[traj_i][i+1] + dt * f(*traj[traj_i][1:]))
        traj.append(x)
        traj_i += 1
        t += dt
    return traj
```

```python
# FitzHugh-Nagumo方程式
fs = [lambda x,y: x - x**3 - y, lambda x,y: x ]
traj = ndsolve(fs, [2,1], 0, 30)

import matplotlib.pyplot as plt
ts, xs, ys = zip(*traj)

plt.plot(xs, ys)
plt.show()
plt.plot(ts, xs)
plt.show()
plt.plot(ts, ys)
plt.show()
```

```python
# Rössler chaos
c1 = 0.15
c2 = 0.2
c3 = 7.1
fs = [lambda s1,s2,s3: -s2 -s3, lambda s1,s2,s3: s1+c1*s2, lambda s1,s2,s3: c2 + s3*(s1-c3) ]
traj = ndsolve(fs, [1, 1, 1], 0, 100)

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Axes3Dでは動作しなかった
ts, xs, ys, zs = zip(*traj)
ax.plot(xs, ys, zs)
plt.show()
```
