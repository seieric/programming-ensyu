```python
def generate_string_from_hash(hash_value, max_len=5):
    string = ""
    for i in range(max_len):
        value = hash_value % 62
        hash_value //= 62
        if value < 10:
            string += str(value)
        elif value < 36:
            string += chr(value - 10 + ord('a'))
        elif value < 62:
            string += chr(value - 36 + ord('A'))
    return string

def generate_short_url(long_url, url_dict):
    hash_value = hash(long_url)
    while True:
        short_url = generate_string_from_hash(hash_value)
        if short_url not in url_dict:
            break
        hash_value += 1
    url_dict[short_url] = long_url
    return short_url

def get_long_url(short_url, url_dict):
    if short_url not in url_dict:
        return None
    else:
        return url_dict[short_url]
```

```python
url_dict = {}
```

```python
short_url = generate_short_url('https://example.com/', url_dict)
print(short_url)
long_url = get_long_url(short_url, url_dict)
print(long_url)
```
