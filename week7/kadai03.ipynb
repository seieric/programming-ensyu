```python
def bubble_sort(array):
    is_swapped = True
    for i in range(len(array)):
        # 直前のループで交換が発生していないのならもう交換するものはないので抜ける
        if not is_swapped:
            break
        is_swapped = False
        for j in range(len(array) - 1 - i):
            if array[j] > array[j + 1]:
                array[j], array[j + 1] = array[j + 1], array[j]
                is_swapped = True
    return array
```

```python
array = [3, 6, 8, 2, 8, 10, 50, 1, 5, 2]
sorted_array = bubble_sort(array)
print(sorted_array)
array = [4, 8, 2, 6]
sorted_array = bubble_sort(array)
print(sorted_array)
```
