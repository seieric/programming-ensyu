```python
import math
```

```python
def is_prime(n):
    if n == 2:
        return True
    if n < 2 or n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n))+1, 2):
        if n % i == 0:
            return False
    return True
```

```python
def euler_product(n):
    res = 1
    for i in range(1, n+1):
        if is_prime(i):
            res *= (1 - (1 / i**2))
    return res
```

```python
pi = math.sqrt(6 / euler_product(100000))
print(pi)
```
